### Cleaning the Data part 1
- Drop columns with empty values exceeding 60%

```python
# target columns
target_columns = ['Effect: Other','Number of people injured','Number of human fatalities','Cost: Aircraft time out of service (hours)',
                  'Location: Freeform en route','Location: Nearby if en route']
# Drop columns
bird_strike_df=bird_strike_df.drop(target_columns, axis=1)
```

### Cleaning Data Part 2
- Converting categorical data to numerical

```python
def mapColumn(feature,df):
  df[feature] = df[feature].map
  return df
```

```python
# replace empty values with Mean - MCT
# Replace empty values with NaN
bird_strike_df.replace('', np.nan, inplace=True)

# Calculate the mean of each column
column_means = bird_strike_df.mean(axis=0)

# Fill missing values with the column means
bird_strike_df = bird_strike_df.fillna(column_means, inplace=True)
```

## Visualization


```python
# Create a scatter chart
plt.scatter(ga_df["Years of Service"], ga_df["Vacation Hours Used"])

# Add a trendline
z = np.polyfit(ga_df["Years of Service"], ga_df["Vacation Hours Used"], 1)
p = np.poly1d(z)
plt.plot(ga_df["Years of Service"], p(ga_df["Years of Service"]), color="red")

# Add labels and title
plt.xlabel("Years of Service")
plt.ylabel("Amount of Vacation Time Used")
plt.title("Scatter Chart with Trendline")

# Show the plot
plt.show()
```

### Differentiating Junior and Senior

```python
# Create the scatter chart
plt.scatter(ga_df['Years of Service'][ga_df['Consultant Title'] == 'Junior'], ga_df['Vacation Hours Used'][ga_df['Consultant Title'] == 'Junior'], label='Junior')
plt.scatter(ga_df['Years of Service'][ga_df['Consultant Title'] == 'Senior'], ga_df['Vacation Hours Used'][ga_df['Consultant Title'] == 'Senior'], label='Senior')

# Add a legend
plt.legend()

# Add labels and title
plt.xlabel('Years of Service')
plt.ylabel('Amount of Vacation Time Used')
plt.title('Scatter Chart by Consultant Level')

# Show the plot
plt.show()
```
